```python
import os
from datasets import load_dataset
import unsloth
from unsloth import FastLanguageModel
from unsloth.chat_templates import standardize_sharegpt
from transformers import AutoTokenizer
from trl import SFTTrainer, SFTConfig
```

```python
# -----------------------
# Config
# -----------------------
MODEL_4BIT = "unsloth/gpt-oss-20b-unsloth-bnb-4bit"  # Unsloth-provided 4-bit model ID
OUTPUT_DIR = "gptoss20b_unsloth_qlora_multilingual"
MAX_SEQ_LEN = 2048
REASONING_EFFORT = "medium"   # low|medium|high
DATASET_ID = "HuggingFaceH4/Multilingual-Thinking"
SPLIT = "train"
```

```python
os.makedirs(OUTPUT_DIR, exist_ok=True)
```

```python
# -----------------------
# Load dataset (datasets lib) — the dataset contains `messages` in Harmony format.
# -----------------------
ds = load_dataset(DATASET_ID, split=SPLIT)
```

```python
ds.data
```

```python
# -----------------------
# Load Unsloth model + tokenizer
# -----------------------
# FastLanguageModel.from_pretrained will configure the model for Unsloth training (4-bit).
model, tokenizer = FastLanguageModel.from_pretrained(
    model_name=MODEL_4BIT,
    max_seq_length=MAX_SEQ_LEN,
    load_in_4bit=True,
    dtype=None,
)
```

```python
# Unsloth helper to add PEFT/LoRA in a QLoRA-friendly way:
model = FastLanguageModel.get_peft_model(
    model,
    r=16,
    lora_alpha=32,
    lora_dropout=0.05,
    bias="none",
    target_modules="all-linear",  # Unsloth picks GPT-OSS linearized modules
    use_gradient_checkpointing=True,
)
```

```python
# -----------------------
# Preprocessing: apply the Harmony chat template using Unsloth tokenizer helpers
# The dataset rows provide a `messages` list. We'll render them to final training text.
# -----------------------
def render_examples(batch):
    out = {"text": []}
    for msgs in batch["messages"]:
        # msgs is already a list of message dicts (developer/user/assistant/analysis/final)
        rendered = tokenizer.apply_chat_template(
            msgs,
            tokenize=False,             # return raw string; trainer will tokenize
            add_generation_prompt=False,
            reasoning_effort=REASONING_EFFORT
        )
        out["text"].append(rendered)
    return out
```

```python

ds_text = ds.map(render_examples, batched=True, remove_columns=[c for c in ds.column_names if c!="messages"])
```

```python
# -----------------------
# Tokenization + SFTTrainer expects the dataset to have a "text" column
# -----------------------
def tokenize_for_trainer(batch):
    tokenized = tokenizer(
        batch["text"],
        truncation=True,
        max_length=MAX_SEQ_LEN,
        padding=False,
    )
    tokenized["labels"] = [ids.copy() for ids in tokenized["input_ids"]]
    return tokenized
```

```python
ds_tok = ds_text.map(tokenize_for_trainer, batched=True, remove_columns=ds_text.column_names)
```

```python
# -----------------------
# SFTTrainer config (TRL)
# -----------------------
sft_cfg = SFTConfig(
    output_dir=OUTPUT_DIR,
    num_train_epochs=1,
    max_steps=-1,
    per_device_train_batch_size=2, # Increased from 1 to 2
    per_device_eval_batch_size=1,
    gradient_accumulation_steps=4,
    logging_steps=20,
    save_steps=200,
    eval_strategy="no",
    eval_steps=200,
    learning_rate=1e-4,
    lr_scheduler_type="cosine",
    warmup_ratio=0.03,
    bf16=False,
    fp16=True,
    gradient_checkpointing=True,
    packing=False,
    dataset_text_field="text",
)
```

```python
trainer = SFTTrainer(
    model=model,
    tokenizer=tokenizer,
    train_dataset=ds_tok,
    eval_dataset=None,   # or pass a split if available
    args=sft_cfg,
)
```
